```python
import matplotlib.pyplot as plt
import numpy as np


```

```python
x = np.linspace(0, 10, 100)
plt.plot(x, x**2)
```

```python
figure = plt.figure()
plt.plot()
```

```python
x = np.linspace(0,2,100)
plt.figure(figsize=(8,6))
plt.plot(x, x, label="linear", color="#43b2a1", linewidth=1.5, linestyle="-.")
plt.plot(x, np.cos(x), label='cosine', color="red", linewidth=2, linestyle=":")
plt.plot(x, np.sin(x), label="sine", color="purple")
plt.xlabel("x label", size=13)
plt.ylabel("y label", size=13)
plt.title("Pyplot simple implementation", size=17)
plt.legend(loc="lower right")
```

```python
fig, ax = plt.subplots(figsize=(8,6))
x = np.linspace(0,2,100)
ax.plot(x, x, color="black", linewidth=1.8, linestyle="-.")
ax.plot(x, np.cos(x),color="black", linewidth=1.8, linestyle="-.")
ax.plot(x, np.sin(x),color="black", linewidth=1.8, linestyle=":")
ax.set_title("Object Oriented Interface", size="14")
ax.legend(["a line (linear)", "b line (cos)", "c line (sin)"])
```

Marking points in visualization

```python
x = np.linspace(0,3,1000)
plt.figure(figsize=(6,4))
plt.plot(x,x**2, x, 2*x)
plt.annotate('x=0', xy=(0,0), xytext=(0,1), arrowprops=dict(facecolor="black", arrowstyle="->"))
plt.annotate('x=2', xy=(2,4), xytext=(2,5), arrowprops=dict(facecolor="black", arrowstyle="->"), bbox=dict(boxstyle="round", alpha=0.2))
```
