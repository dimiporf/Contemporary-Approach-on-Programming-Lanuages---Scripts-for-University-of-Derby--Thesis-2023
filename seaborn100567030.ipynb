PENGUIN DATASET

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

CAN WE TELL APART?

```python
grid = sns.pairplot(data=penguins, hue='species', corner=True)
grid.fig.suptitle('Feature Comparison in penguin species', size=24)
plt.tight_layout()
grid.fig.subplots_adjust(top=0.9)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

grid = sns.pairplot(data=penguins, hue='species', corner=True)
grid.fig.suptitle('Feature Comparison in Penguin Species', size=24)
plt.tight_layout()
grid.fig.subplots_adjust(top=0.9)

# Customize the plot appearance
grid.map_offdiag(sns.scatterplot, s=50)
grid.map_diag(sns.histplot, kde=True, color='skyblue')
grid.add_legend(title='Species', label_order=['Adelie', 'Chinstrap', 'Gentoo'])

# Adjust the tick labels rotation for better readability
plt.xticks(rotation=45)
plt.yticks(rotation=45)

# Show the plot
plt.show()
```

TELL SPECIES APART USING JOINT PLOTS

```python
penguins.dropna(inplace=True)

# Add 'demography' column to penguins DataFrame
penguins['demography'] = penguins.apply(lambda row: "%s_%s" % (row['species'], row['sex']), axis=1)
```

```python
penguins.head()
```

```python
sns.jointplot(data=penguins, x='bill_length_mm', y='body_mass_g', hue='demography', kind='scatter', palette='hls')
```

```python
sns.jointplot(data=penguins, x='bill_length_mm', y='body_mass_g', hue='demography', kind='kde', alpha=0.5, fill=True, height=8, palette='viridis')
```

```python
joint_grid = sns.jointplot(data=penguins, x="bill_length_mm", y="body_mass_g", hue="demography",palette="colorblind" ,height=8)

# Customize the plot appearance
joint_grid.plot_joint(sns.kdeplot, fill=True, alpha=0.4)
joint_grid.ax_marg_x.set_axis_off()
joint_grid.ax_marg_y.set_axis_off()

# Add a legend with a customized title
joint_grid.ax_joint.legend(title='Demography')

# Set title and axis labels
joint_grid.ax_joint.set_xlabel('Bill Length (mm)', size=14)
joint_grid.ax_joint.set_ylabel('Body Mass (g)', size=14)

# Set title above the plot
plt.suptitle('Bill Length vs Body Mass', size=18)

# Show the plot
plt.show()
```
